```python
insc_por_regiao  = pda_prouni.groupby(by='REGIAO_BENEFICIARIO_BOLSA').REGIAO_BENEFICIARIO_BOLSA.count()
```

```python
insc_por_regiao
```

```python
data = [go.Bar(x=insc_por_regiao.index,
               y=insc_por_regiao.values,
               marker = {'color':'lightblue',
                         'line':{'color':'#333',
                                  'width':2}
                        },
               opacity=0.7
              )
                    
       ]

configuracoes_layout = go.Layout(title='INSCRIÇÃO POR REGIÃO',
                               yaxis={'title':'Qtd de inscritos por região'},
                               xaxis={'title':'Região'})


fig = go.Figure(data=data, layout=configuracoes_layout)

#py.iplot(fig)
pyo.plot(fig)
```

```python
pda_prouni['SEXO_BENEFICIARIO_BOLSA'] = pda_prouni['SEXO_BENEFICIARIO_BOLSA'].apply(lambda x: x.upper())
```

```python
for index, row in pda_prouni.iterrows():
    if row.SEXO_BENEFICIARIO_BOLSA == 'F':
        pda_prouni.at[index,'SEXO_BENEFICIARIO_BOLSA'] = 'FEMININO' 
```

```python
for index, row in pda_prouni.iterrows():
    if row.SEXO_BENEFICIARIO_BOLSA == 'M':
        pda_prouni.at[index,'SEXO_BENEFICIARIO_BOLSA'] = 'MASCULINO' 
```
